```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df_ = pd.read_csv("./GroceryStoreDataSet.csv",delimiter=',')
df_.columns = ['Items']
df_['Transaction'] = np.arange(1,df_.shape[0]+1)
df_
```

```python
dict_ = {}
for transaction,items in zip(df_['Transaction'],df_['Items']):
    dict_[transaction] = list(items.split(','))
dict_
```

```python
# create a pandas DataFrame with the above dict
transaction = list()
item = list()
for key in dict_:
    for value in dict_[key]:
        transaction.append(key)
        item.append(value)
print(item)
dict_data = {}
dict_data = {'Transaction' : transaction, 'Items' : item}
```

```python
df = pd.DataFrame(dict_data)
df
```

```python
df.loc[df['Items']=='NONE',:]
```

```python
df.drop(df.loc[df['Items']=='NONE',:].index,axis=0,inplace=True)
```

```python
df.shape
```

```python
hot_encoded_df = df.groupby(['Transaction','Items'])['Items'].count().unstack().reset_index().fillna(0).set_index('Transaction')
```

```python
hot_encoded_df
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    else:
        return 1
hot_encoded_df = hot_encoded_df.applymap(encode_units)
```

```python
hot_encoded_df
```

```python
freq_itemsets = apriori(hot_encoded_df, min_support=0.1,use_colnames = True)
```

```python
freq_itemsets
```

```python
rules = association_rules(freq_itemsets, metric='lift',min_threshold=0.2)
rules
```

```python
rules[(rules['confidence'] > 0.4)]
```
